```python
# Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Data Exploration and Processing


# Load dataset
df = pd.read_csv("aerofit_treadmill.csv")

# Display first few rows
df.head()

```

```python
# Check the shape of the dataset
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
# Check data types of each column
df.info()

# Check for missing values
print("\nMissing values in each column:\n", df.isnull().sum())
```

```python
# Check for duplicate rows
print(f"Duplicate Rows: {df.duplicated().sum()}")
```

```python
# Display first few rows of the dataset again to review data
df.head()
```

```python
# 2. Statistical Summary


#Statistical summary for numerical features
print("Statistical Summary for Numerical Features:")
print(df.describe())

# Statistical summary for categorical features
print("\nStatistical Summary for Categorical Features:")
print(df.describe(include=['object']))
```

```python
# Non Graphical Analysis


# Value counts for all categorical features
print("Value Counts for Product:\n", df['Product'].value_counts())
print("\nValue Counts for Gender:\n", df['Gender'].value_counts())
print("\nValue Counts for MaritalStatus:\n", df['MaritalStatus'].value_counts())
```

```python
# Unique values for all categorical features
print("Unique values in Product:", df['Product'].unique())
print("Unique values in Gender:", df['Gender'].unique())
print("Unique values in MaritalStatus:", df['MaritalStatus'].unique())
```

```python
# 4. Graphical Analysis


# Univariate Analysis - Numerical features:

# Distribution plot for numerical features (Age, Income, Miles)
plt.figure(figsize=(12, 8))

# Distribution plot for Age
plt.subplot(2, 2, 1)
sns.histplot(df['Age'], kde=True, color='lightblue')  # Light Blue color
plt.title('Distribution of Age')

# Distribution plot for Income
plt.subplot(2, 2, 2)
sns.histplot(df['Income'], kde=True, color='lightgreen')  # Light Green color
plt.title('Distribution of Income')

# Distribution plot for Miles
plt.subplot(2, 2, 3)
sns.histplot(df['Miles'], kde=True, color='lightpink')  # Light Pink color
plt.title('Distribution of Miles')

plt.tight_layout()
plt.show()

# Count plot for Usage (discrete variable)
plt.figure(figsize=(6, 4))
sns.countplot(x=df['Usage'], palette='pastel')  # Pastel color palette for lighter shades
plt.title('Count Plot of Usage (Times per Week)')
plt.xlabel('Usage (times per week)')
plt.ylabel('Count')
plt.show()

# Box plots for numerical features (Age, Income, Usage, Miles)
plt.figure(figsize=(12, 8))

# Box plot for Age
plt.subplot(2, 2, 1)
sns.boxplot(x=df['Age'], color='lightblue')  # Light Blue color
plt.title('Box Plot of Age')

# Box plot for Income
plt.subplot(2, 2, 2)
sns.boxplot(x=df['Income'], color='lightgreen')  # Light Green color
plt.title('Box Plot of Income')

# Box plot for Usage (discrete variable)
plt.subplot(2, 2, 3)
sns.boxplot(x=df['Usage'], color='lightcoral')  # Light Coral color
plt.title('Box Plot of Usage')

# Box plot for Miles
plt.subplot(2, 2, 4)
sns.boxplot(x=df['Miles'], color='lightpink')  # Light Pink color
plt.title('Box Plot of Miles')

plt.tight_layout()
plt.show()

```

```python
# Univariate Analysis - Categorical features


 #Count plot for Product
plt.figure(figsize=(6, 4))
sns.countplot(x='Product', data=df, palette='viridis')
plt.title('Count of Products Purchased')
plt.show()

# Count plot for Gender
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=df, palette='viridis')
plt.title('Count of Customers by Gender')
plt.show()

# Count plot for Marital Status
plt.figure(figsize=(6, 4))
sns.countplot(x='MaritalStatus', data=df, palette='viridis')
plt.title('Count of Customers by Marital Status')
plt.show()
```

```python
# Bivariate Analysis

# Product vs Gender
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Product', hue='Gender', palette='pastel')
plt.title('Product vs Gender')
plt.xlabel('Product')
plt.ylabel('Count')
plt.show()

# Product vs MaritalStatus
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Product', hue='MaritalStatus', palette='pastel')
plt.title('Product vs MaritalStatus')
plt.xlabel('Product')
plt.ylabel('Count')
plt.show()

# Product vs Age (Box Plot)
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Product', y='Age', palette='pastel')
plt.title('Product vs Age')
plt.xlabel('Product')
plt.ylabel('Age')
plt.show()
```

```python
# Multivariate Analysis

# Selecting numerical columns for pairplot
numerical_features = ['Age', 'Income', 'Usage', 'Miles']

# Create the pairplot with 'Product' as hue
sns.pairplot(df[numerical_features + ['Product']], diag_kind='kde', hue='Product', palette='pastel')

# Display the plot
plt.show()


```

```python
# 5. Correlation Analysis



# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Calculate correlation matrix for numeric columns
corr_matrix = numeric_df.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# 6. Outlier Detection



# Calculate Q1 (25th percentile) and Q3 (75th percentile) for each numerical column
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR)))

# Display the outliers
print("Outliers Detected in the Dataset:")
print(outliers.sum())
```

```python
# 7. Conditional Probabilities


# 7.1 What percent of customers have purchased KP281, KP481, or KP781?

# Total number of customers
total_customers = df.shape[0]

# Count the number of customers who purchased KP281
purchased_KP281 = df[df['Product'] == 'KP281']
percentage_KP281 = (purchased_KP281.shape[0] / total_customers) * 100

# Count the number of customers who purchased KP481
purchased_KP481 = df[df['Product'] == 'KP481']
percentage_KP481 = (purchased_KP481.shape[0] / total_customers) * 100

# Count the number of customers who purchased KP781
purchased_KP781 = df[df['Product'] == 'KP781']
percentage_KP781 = (purchased_KP781.shape[0] / total_customers) * 100

# Print the results
print(f"Percentage of Customers Purchasing KP281: {percentage_KP281:.2f}%")
print(f"Percentage of Customers Purchasing KP481: {percentage_KP481:.2f}%")
print(f"Percentage of Customers Purchasing KP781: {percentage_KP781:.2f}%")

```

```python
# 7.2 Create frequency tables and calculate the percentage


# Total number of customers
total_customers = df.shape[0]

# 1. Percentage of Male Customers Purchasing a Treadmill (any treadmill model)
male_customers = df[df['Gender'] == 'Male']
male_customers_purchasing_treadmill = male_customers[male_customers['Product'].isin(['KP281', 'KP481', 'KP781'])]
percentage_male_customers = (male_customers_purchasing_treadmill.shape[0] / total_customers) * 100
print(f"Percentage of Male Customers Purchasing a Treadmill: {percentage_male_customers:.2f}%")

# 2. Percentage of Female Customers Purchasing KP781
female_customers_KP781 = df[(df['Gender'] == 'Female') & (df['Product'] == 'KP781')]
total_KP781_customers = df[df['Product'] == 'KP781']
percentage_female_KP781 = (female_customers_KP781.shape[0] / total_KP781_customers.shape[0]) * 100
print(f"Percentage of Female Customers Purchasing KP781: {percentage_female_KP781:.2f}%")

# 3. Probability of a Customer Being Female Given That the Product is KP281
female_KP281 = df[(df['Gender'] == 'Female') & (df['Product'] == 'KP281')]
total_KP281 = df[df['Product'] == 'KP281']
probability_female_KP281 = (female_KP281.shape[0] / total_KP281.shape[0]) * 100
print(f"Probability of a Customer Being Female Given Product is KP281: {probability_female_KP281:.2f}%")
```

```python
# Product – Age
# Percentage of customers with Age between 20s and 30s among all customers
age_range_percentage = (df[(df['Age'] >= 20) & (df['Age'] <= 30)]['Product'].count() / df['Product'].count()) * 100
print(f"Percentage of customers with Age between 20s and 30s: {age_range_percentage:.2f}%")
```

```python
# Product – Income

# Percentage of low-income customers purchasing a treadmill
low_income_percentage = (df[df['Income'] < 30000]['Product'].count() / df['Product'].count()) * 100

# Percentage of high-income customers purchasing KP781
high_income_kp781_percentage = (df[(df['Income'] >= 100000) & (df['Product'] == 'KP781')].shape[0] / df[df['Income'] >= 100000].shape[0]) * 100

# Percentage of high-income customers purchasing KP781 given that Product is KP781
prob_high_income_given_kp781 = (df[(df['Income'] >= 100000) & (df['Product'] == 'KP781')].shape[0] / df[df['Product'] == 'KP781'].shape[0]) * 100

print(f"Percentage of low-income customers purchasing a treadmill: {low_income_percentage:.2f}%")
print(f"Percentage of high-income customers purchasing KP781: {high_income_kp781_percentage:.2f}%")
print(f"Percentage of high-income customers purchasing KP781 given that Product is KP781: {prob_high_income_given_kp781:.2f}")
```

```python
# Product – Fitness

# Percentage of customers with Fitness Level 5
fitness_level_5_percentage = (df[df['Fitness'] == 5]['Product'].count() / df['Product'].count()) * 100

# Percentage of customers with Fitness Level 5 purchasing KP781
fitness_level_5_kp781_percentage = (df[(df['Fitness'] == 5) & (df['Product'] == 'KP781')].shape[0] / df[df['Fitness'] == 5].shape[0]) * 100

# Percentage of Fitness Level 5 customers buying KP781
fitness_level_5_buy_kp781_percentage = (df[(df['Fitness'] == 5) & (df['Product'] == 'KP781')].shape[0] / df.shape[0]) * 100

print(f"Percentage of customers with Fitness Level 5: {fitness_level_5_percentage:.2f}%")
print(f"Percentage of customers with Fitness Level 5 purchasing KP781: {fitness_level_5_kp781_percentage:.2f}%")
print(f"Percentage of Fitness Level 5 customers buying KP781: {fitness_level_5_buy_kp781_percentage:.2f}%")
```

```python
# Product- Marital Status
# Total number of customers
total_customers = df.shape[0]

# Filter customers who are partnered
partnered_customers = df[df['MaritalStatus'] == 'Partnered']

# Filter partnered customers who purchased any treadmill (KP281, KP481, KP781)
partnered_customers_purchasing_treadmills = partnered_customers[partnered_customers['Product'].isin(['KP281', 'KP481', 'KP781'])]

# Calculate the percentage of partnered customers purchasing treadmills
percentage_partnered_customers = (partnered_customers_purchasing_treadmills.shape[0] / total_customers) * 100

print(f"Percentage of Customers who are Partnered Purchasing Treadmills: {percentage_partnered_customers:.2f}%")
```
